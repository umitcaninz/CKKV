## Entropi Yöntemi Nedir?
Entropi yöntemi, çok kriterli karar verme (ÇKKV) problemlerinde kriterlerin ağırlıklarını objektif bir şekilde hesaplamak için kullanılan bir yöntemdir. Temel olarak, bir sistemdeki düzensizlik ve rastgelelik ölçüsünü kullanan termodinamikten türemiştir.

#### Entropi Yönteminin Kullanım Amacı
Entropi yöntemi, ÇKKV problemlerinde aşağıdaki amaçlar için kullanılır:

##### - Kriterlerin objektif bir şekilde ağırlıklandırılması
##### - Karar verme sürecinde tüm kriterlerin dikkate alınması
##### - Kararların daha tutarlı ve doğru olması
##### - Entropi Yönteminin Kullanıldığı Sektörler

### Entropi yöntemi, birçok farklı sektörde ÇKKV problemlerini çözmek için kullanılır. Başlıca kullanım alanları şunlardır:

##### - Mühendislik: Ürün tasarımı, malzeme seçimi, proses optimizasyonu
##### - Finans: Yatırım seçimi, risk analizi, portföy optimizasyonu
##### - Yönetim: Personel seçimi, proje değerlendirmesi, tedarikçi seçimi
##### - Sağlık: Hasta teşhisi, tedavi seçimi, kaynak tahsisi
##### - Enerji: Enerji kaynağı seçimi, tesis tasarımı, şebeke optimizasyonu
## Entropi Yönteminin Avantajları
##### Entropi yönteminin diğer kriter ağırlıklandırma yöntemlerine göre birçok avantajı vardır:

##### - Objektif: Kriterlerin ağırlıkları, karar vericinin önyargılarından etkilenmeden hesaplanır.
##### - Veriye Dayalı: Kriterlerin ağırlıkları, mevcut verilere dayalı olarak hesaplanır.
##### - Esnek: Farklı tipteki kriterlere uygulanabilir.
##### - Kolay Kullanım: Hesaplaması kolaydır ve yaygın olarak kullanılan yazılımlarla uygulanabilir.

### Entropi Yönteminin Dezavantajları
Entropi yönteminin bazı dezavantajları da vardır:

##### - Veriye Bağımlılık: Yöntemin doğruluğu, kullanılan verilerin kalitesine ve doğruluğuna bağlıdır.
##### - Hesaplama Karmaşıklığı: Büyük veri kümeleri için hesaplama karmaşıklığı artabilir.
##### - Kriterlerin Bağımsızlığı: Kriterlerin birbirinden bağımsız olması varsayılır, bu her zaman doğru olmayabilir.

### Entropi Yönteminin Sonucu
##### - Entropi yöntemi sonucunda her bir kriter için bir ağırlık değeri elde edilir. Bu değerler, karar verme sürecinde kriterlerin ne kadar önemli olduğunu gösterir.

### Elde Edilen Çıktının Yorumlanması
##### - Elde edilen çıktıdaki ağırlık değerleri, yüzde veya oran olarak yorumlanabilir. En yüksek ağırlık değerine sahip kriter, en önemli kriter olarak kabul edilir.

### Örnek
Diyelim ki bir şirket yeni bir ürün geliştirmek için üç farklı malzemeyi karşılaştırıyor. Her bir malzeme için üç kriter dikkate alınıyor:

##### - Maliyet
##### - Dayanıklılık
##### - Estetik

#### Entropi yöntemi kullanılarak her bir kriter için bir ağırlık değeri hesaplanıyor:

##### - Maliyet: %40
##### - Dayanıklılık: %30
##### - Estetik: %30
#### Bu sonuçlara göre, karar verme sürecinde maliyet en önemli kriter olarak kabul edilir.

### Entropi Yöntemini Kullanarak Yurt Seçimi Python Uygulaması

##### Bir öğrencisiniz ve Ankara'da okumak için yurt arıyorsunuz. Aşağıdaki kriterlere göre size en uygun yurdu seçmek istiyorsunuz:

#####  Fiyat: 1000 TL ile 1500 TL arasında
#####  Uzaklık: Üniversiteye 100 metre ile 300 metre arasında
#####  Oda Tipi: Tek Kişilik, İki Kişilik veya Üç Kişilik
##### Yemekhane: Var veya Yok
##### Güvenlik: İyi, Orta veya Kötü

#####  Entropi yöntemini kullanarak size en uygun yurdu hangi değişkene dikkat ederek seçmelisiniz ?

```python
import random
from sklearn.preprocessing import LabelEncoder
import math
import numpy as np
import pandas as pd
```

```python
fiyat_min = 1000
fiyat_max = 1500
uzaklik_min = 100
uzaklik_max = 300

yurtlar = []

for i in range(50):
    yurt =  f"Yurt_{i+1}"
    fiyat = random.randint(fiyat_min, fiyat_max)
    uzaklik = random.randint(uzaklik_min, uzaklik_max)
    oda_tipi = random.choice(["Tek Kişilik", "İki Kişilik","Üç Kişilik"])
    yemekhane = random.choice(["Var", "Yok"])
    guvenlik = random.choice(["İyi", "Orta", "Kötü"])

    # Yurt bilgilerini bir sözlükte saklamak
    yurt = {
        "Yurt":yurt,
        "Fiyat": fiyat,
        "Uzaklık": uzaklik,
        "Oda Tipi": oda_tipi,
        "Yemekhane": yemekhane,
        "Güvenlik": guvenlik,
    }

    # Yurt sözlüğünü listeye eklemek
    yurtlar.append(yurt)
```

```python
yurtlar = pd.DataFrame(yurtlar)
```

```python
yurtlar
```

```python
le = LabelEncoder()
```

```python
yurtlar["Oda Tipi"] = le.fit_transform(yurtlar["Oda Tipi"])
yurtlar["Yemekhane"] = le.fit_transform(yurtlar["Yemekhane"])
yurtlar["Güvenlik"] = le.fit_transform(yurtlar["Güvenlik"])
```

```python
yurtlar
```

```python
for col in yurtlar.columns:
    if col != "Yurt":
        yurtlar[col] = yurtlar[col] / yurtlar[col].sum()

# Normalize edilmiş veri setini kaydedin
normalize_df = yurtlar.copy()
```

```python
normalize_df["Fiyat"].sum()
```

```python
normalize_df
```

```python
entropies = {}
n = len(yurtlar)
k = 1 / np.log(n)

for column in yurtlar.columns[1:]:  # Yurt sütunu hariç tüm sütunları hesaplamak için [1:] kullanıldı.
    p = yurtlar[column].value_counts(normalize=True)  # Olayların olasılıklarını hesapla
    entropy = -k * np.sum(p * np.log(p))  # Entropi formülü
    entropies[column] = entropy

print("Entropi Değerleri:")
for column, entropy in entropies.items():
    print(f"{column}: {entropy}")
```

```python
for column, entropy in entropies.items():
    print(f"{column}: {1-entropy}")
```

```python
total = sum([0.014174705610844662, 0.028349411221689325, 0.7230203593936051,0.8246611026313987, 0.7192732407684302])

ratios = {
    'Fiyat': 0.014174705610844662 / total,
    'Uzaklık':  0.028349411221689325 / total,
    'Oda Tipi': 0.7230203593936051 / total,
    'Yemekhane': 0.8246611026313987 / total,
    'Güvenlik': 0.7192732407684302 / total
}

print("Değişken Ağırlıkları:")
for column, ratio in ratios.items():
    print(f"{column}: {ratio}")
```
