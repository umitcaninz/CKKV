<h1 style="font-weight: bold; color: green;">Simple Additive Weighting (SAW) Modeli</h1>

<h2 style="font-weight: bold; color: green;">Adımlar:</h2>
<p style="font-weight: bold; font-size: 15px;"> 1. Kriterlerin belirlenmesi</p>
<p style="font-weight: bold; font-size: 15px;">2. Her kriter için ağırlıkların belirlenmesi</p>
<p style="font-weight: bold; font-size: 15px;">3. Alternatiflerin her kritere göre puanlanması</p>
<p style="font-weight: bold; font-size: 15px;">4. Puanların kriter ağırlıklarıyla çarpılması ve toplanması</p>
<p style="font-weight: bold; font-size: 15px;">5. En yüksek toplam puana sahip olan alternatifin seçilmesi</p>

<p style="font-weight: bold; font-size: 20px;color: purple;">Simple Additive Weighting (SAW) modeli, T.L. Saaty tarafından geliştirilmiştir.</p>

<h2 style="font-weight: bold; color: green;">Avantajları:</h2>
<p style="font-weight: bold; font-size: 15px;">- Kolay anlaşılabilir ve uygulanabilir.</p>
<p style="font-weight: bold; font-size: 15px;">- Basit matematiksel işlemlerle sonuç elde edilir.</p>
<p style="font-weight: bold; font-size: 15px;"> - Kullanıcı dostudur ve esneklik sağlar.</p>

<h3 style="font-weight: bold; color: green;">Sonuçlar Nasıl Yorumlanır ? :</h3>
<p style="font-weight: bold; font-size: 15px;">SAW modeli sonuçları, toplam puanı en yüksek olan alternatifin en iyi seçenek olduğunu gösterir. Bu alternatif, verilen kriterlere göre diğer alternatiflere göre daha uygun olarak değerlendirilmiştir.</p>


```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
veriler = {
    "Tedarikçiler": ["Tedarikçi A", "Tedarikçi B", "Tedarikçi C", "Tedarikçi D", "Tedarikçi E",
           "Tedarikçi F", "Tedarikçi G", "Tedarikçi H", "Tedarikçi I", "Tedarikçi J",
           "Tedarikçi K", "Tedarikçi L", "Tedarikçi M", "Tedarikçi N", "Tedarikçi O",
           "Tedarikçi P", "Tedarikçi Q", "Tedarikçi R", "Tedarikçi S", "Tedarikçi T"],
    
    "Kalite": [8, 3, 3, 5, 5, 2, 5, 5, 1, 1, 3, 8, 6, 7, 5, 2, 1, 7, 9, 3],
    
    "Fiyat": [2, 2, 5, 8, 9, 9, 4, 3, 9, 5, 3, 9, 9, 2, 2, 5, 1, 3, 7, 7],
    
    "Teslim Süresi": [4, 8, 3, 4, 4, 8, 9, 8, 6, 6, 4, 1, 2, 6, 6, 6, 4, 1, 8, 5]
}

df = pd.DataFrame(veriler)
df.set_index("Tedarikçiler", inplace=True)
```

```python
# Ağırlıklar
agirliklar = [0.4, 0.3, 0.3]

# SAW puanlarını hesapla
df['SAW Puanı'] = df.apply(lambda row: sum(row * agirliklar), axis=1)

df_sorted = df.sort_values(by='SAW Puanı', ascending=False)
df_sorted
```

```python
df_sorted = df.sort_values(by='SAW Puanı', ascending=False)
plt.figure(figsize=(10, 6))
df_sorted['SAW Puanı'].plot(kind='bar', color='green')
plt.title('Tedarikçilerin SAW Puanları (Büyükten Küçüğe)')
plt.xlabel('Tedarikçiler')
plt.ylabel('SAW Puanı')
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

<p style="font-weight: bold; font-size: 17px;color: green;">"Tedarikçi S" en yüksek SAW puanına sahiptir (8.1). Yüksek kalite ve rekabetçi fiyatları ile tercih edilen bir seçenek olabilir.</p>
<p style="font-weight: bold; font-size: 17px;color: green;">"Tedarikçi L" (6.2) ve "Tedarikçi F" (5.9), "Tedarikçi S" den sonra gelen diğer iyi seçeneklerdir.</p>
<p style="font-weight: bold; font-size: 17px;color: green;">"Tedarikçi Q" en düşük SAW puanına sahiptir (1.9). Bu tedarikçi, diğerlerine göre daha düşük kalite ve/veya daha yüksek fiyatlar veya teslim süreleri sunuyor olabilir.</p>
<p style="font-weight: bold; font-size: 17px;color: green;">Bu yorumlar, işletmenin tercihlerine ve önceliklerine bağlı olarak değişebilir. Örneğin, bir işletme düşük fiyatları tercih ediyorsa, daha düşük SAW puanına sahip bir tedarikçi bile tercih edilebilir.</p>
