<h1 style="color: skyblue;">PROMETHEE Nedir?</h1>
<p style="font-size: 15px; font-weight: bold; color: black;">Preference Ranking Organization Method for Enrichment Evaluation (PROMETHEE), Bernard Roy tarafından 1960'ların sonlarında geliştirilen bir çok kriterli karar verme yöntemidir. PROMETHEE, alternatifler arasında tercih sıralaması yapmak için kullanılır ve karar verme sürecindeki belirsizlikleri azaltmayı, karmaşıklığı yönetmeyi ve net sonuçlar sunmayı hedefler.</p>

<h2 style="color: skyblue;">PROMETHEE Yöntemi ve Adımları</h2>
<ul style="font-size: 15px; color: black;">
  <li>Ağırlıklandırma ve Normalizasyon</li>
  <li>Akış Matrisinin Oluşturulması</li>
  <li>Kriterlerin Karşılaştırılması</li>
  <li>PROMETHEE Endeksi Hesaplama</li>
  <li>Sonuçların Sunumu</li>
</ul>

<h2 style="color: skyblue;">PROMETHEE'nin Avantajları</h2>
<ul style="font-size: 15px; color: black;">
  <li>Yapısal olarak basit ve anlaşılması kolaydır.</li>
  <li>Çok kriterli karar verme problemlerini ele almak için etkilidir.</li>
  <li>Karar vericinin tercihlerini ve kriterlerin önem derecelerini dikkate alır.</li>
</ul>

<h2 style="color: skyblue;">AHP ile Farkları</h2>
<ul style="font-size: 15px; color: black;">
  <li>AHP, karar verme sürecinde karar alıcıların tercihlerini hiyerarşik bir yapı içinde belirlemek için kullanılırken, PROMETHEE direkt olarak alternatifler arasında sıralama yapmak için kullanılır.</li>
  <li>AHP, karar vericinin tercihlerini ölçeklendirme ve ağırlıklandırma sürecinde kullanılırken, PROMETHEE'de ağırlıklar genellikle önceden belirlenir veya analitik yöntemlerle belirlenir.</li>
</ul>

<h2 style="color: skyblue;">PROMETHEE'nin Objektifliği ve Subjektifliği</h2>
<p style="font-size: 15px; font-weight: bold; color: black;">PROMETHEE'nin objektifliği ve subjektifliği, yöntemin kullanım amacına ve uygulama bağlamına bağlıdır. PROMETHEE'nin temel amacı, belirli bir karar verme probleminde alternatifleri objektif bir şekilde sıralamaktır. Ancak, karar vericinin öznellikleri ve tercihleri yöntemi uygulama sürecinde önemli bir rol oynar.</p>

<h2 style="color: skyblue;">PROMETHEE'nin Kullanım Alanları</h2>
<p style="font-size: 15px; font-weight: bold; color: black;">PROMETHEE yöntemi, geniş bir uygulama alanına sahiptir ve endüstriyel, ekonomik, çevresel ve sosyal karar verme süreçlerinde kullanılabilir. Özellikle karmaşık ve çok kriterli karar verme problemlerinde etkili bir araç olarak kabul edilir.</p>


<h2 style="color: skyblue;">ÖRNEK</h2>


<p style="font-size: 15px; font-weight: bold; color: black;">Şehirde yeni bir restoran açma konusunda farklı bölgeler arasında karar vermemiz gerekiyor. Bu kararı vermek için PROMETHEE yöntemini kullanacağız ve farklı kriterleri (nüfus yoğunluğu, rekabet durumu, ulaşılabilirlik vb.) dikkate alacağız.</p>

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

bölgeler = ['Bölge {}'.format(i) for i in range(1, 21)]

kriterler = ['Nüfus Yoğunluğu', 'Rekabet Durumu', 'Ulaşılabilirlik']

np.random.seed(0)
performans_verisi = np.random.randint(1, 10, size=(20, 3))

df = pd.DataFrame(data=performans_verisi, columns=kriterler, index=bölgeler)
```

```python
df
```

```python
weights = np.array([1/len(kriterler)] * len(kriterler))

def karşılaştırma_fonksiyonu(x, y):
    return x - y

akış_matrisi = pd.DataFrame(index=bölgeler, columns=bölgeler, dtype=float)

for i in range(len(bölgeler)):
    for j in range(len(bölgeler)):
        if i != j:
            akış_matrisi.iloc[i, j] = np.sum(karşılaştırma_fonksiyonu(df.iloc[i], df.iloc[j]) * weights)

PROMETHEE_endeksi = akış_matrisi.sum(axis=1)

sıralama = PROMETHEE_endeksi.sort_values(ascending=False)
print("PROMETHEE Sıralaması:")
print(sıralama)

```

<h3 style="color: skyblue;">En İyi Bölgeler (Pozitif Endeksler):</h3>
<p style="font-size: 15px; font-weight: bold; color: black;">Bölge 6, en yüksek PROMETHEE endeksine sahiptir (67.33). Bu, diğer bölgelere göre en üstün performansı sergilediği anlamına gelir. Bölge 4, 17 ve 20 de diğer bölgelere göre yüksek performans gösteren bölgelerdir.</p>

<h3 style="color: skyblue;">Orta Seviye Bölgeler (Yaklaşık Sıfır Endeks):</h3>
<p style="font-size: 15px; font-weight: bold; color: black;">Bölge 14, PROMETHEE endeksi 0.67 olan tek bölgedir. Bu, diğer bölgelere göre yaklaşık olarak eşit bir performans sergilediği anlamına gelir.</p>

<h3 style="color: skyblue;">En Zayıf Bölgeler (Negatif Endeksler):</h3>
<p style="font-size: 15px; font-weight: bold; color: black;">Bölge 13, en düşük PROMETHEE endeksine sahiptir (-46.00). Bu, diğer bölgelere göre en zayıf performansı sergilediği anlamına gelir. Bölge 11, 8, 9 ve 1 de diğer bölgelere göre zayıf performans gösteren bölgelerdir.</p>


```python
plt.figure(figsize=(10, 6))
plt.barh(sıralama.index, sıralama.values, color='skyblue')
plt.xlabel('PROMETHEE Endeksi')
plt.ylabel('Bölge')
plt.title('PROMETHEE Sonuçları')
plt.gca().invert_yaxis()  # Bölgeleri büyükten küçüğe sıralamak için y eksenini tersine çevirir
plt.show()
```
